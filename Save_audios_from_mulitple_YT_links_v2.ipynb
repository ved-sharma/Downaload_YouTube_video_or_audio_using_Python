```python
from pytubefix import YouTube
```

```python
YT_links = ["https://www.youtube.com/watch?v=cUuEPZEGpIk",
            "https://www.youtube.com/watch?v=v2DerSDbfCg",
            ]
```

```python
yt = [YouTube(link) for link in YT_links]

# print the title of the videos
for video in yt:
    print(video.title)
```

```python
!pwd
destination = '.' # download audio files to the current directory
```

```python
for video in yt:
    # stream = video.streams.filter(only_audio=True).first()
    stream = video.streams.get_audio_only()
    stream.download(output_path=destination)

print("All audio files downloaded successfully!")
```
